```python
import pandas as pd
import scipy.stats
import numpy as np
start_time = pd.datetime.now()
print(start_time)
```

Read & Inspect Data

```python
%%time
base = pd.read_excel('card transactions.xlsx')
```

```python
card = base
card.head(30)
```

```python
card.mode()
```

Remove Outlier and all but type P

```python
card = card[card.Transtype == 'P']
card = card[card.Recnum != 52715]
```

Fill Missing Values

```python
state_ref = card.groupby('Merch zip')['Merch state'].apply(lambda x: x.mode()).reset_index()
state_ref = state_ref[state_ref.level_1 == 0].iloc[:,[0,2]]
```

```python
state_ref.columns = ['Merch zip', 'state ref']
```

```python
card = card.merge(state_ref, on = 'Merch zip', how = 'left')
```

```python
card['Merch state'] = card['Merch state'].fillna(card['state ref']).fillna('TN')
```

```python
card['Merch state'].isnull().sum()
```

```python
zip_ref = card.groupby(['Merch state', 'Cardnum'])['Merch zip'].apply(lambda x: x.mode()).reset_index()
zip_ref = zip_ref[zip_ref.level_2 == 0].iloc[:,[0,1,3]]
```

```python
zip_ref.columns = ['Merch state', 'Cardnum', 'zip ref']
```

```python
card = card.merge(zip_ref, on = ['Merch state', 'Cardnum'], how = 'left')
```

```python
zip_ref2 = card.groupby(['Merch state'])['Merch zip'].apply(lambda x: x.mode()).reset_index()
zip_ref2 = zip_ref2[zip_ref2.level_1 == 0].iloc[:,[0,2]]
zip_ref2.columns = ['Merch state', 'zip ref2']
```

```python
card = card.merge(zip_ref2, on = 'Merch state', how = 'left')
```

```python
card['Merch zip']= card['Merch zip'].fillna(card['zip ref']).fillna(card['zip ref2']).fillna(38118)
```

```python
card['Merch zip'].isnull().sum()
```

```python
merch_ref = card.groupby(['Merch state', 'Cardnum'])['Merchnum'].apply(lambda x: x.mode()).reset_index()
merch_ref = merch_ref[merch_ref.level_2 == 0].iloc[:,[0,1,3]]
```

```python
merch_ref.columns = ['Merch state', 'Cardnum', 'merch ref']
```

```python
card = card.merge(merch_ref, on = ['Merch state', 'Cardnum'], how = 'left')
```

```python
merch_ref2 = card.groupby(['Merch state'])['Merchnum'].apply(lambda x: x.mode()).reset_index()
merch_ref2 = merch_ref2[merch_ref2.level_1 == 0].iloc[:,[0,2]]
merch_ref2.columns = ['Merch state', 'merch ref2']
```

```python
card = card.merge(merch_ref2, on = 'Merch state', how = 'left')
```

```python
card['Merchnum']= card['Merchnum'].fillna(card['merch ref']).fillna(card['merch ref2']).fillna(930090121224)
```

```python
card['Merchnum'].isnull().sum()
```

Amount Variables & Frequency Variables

```python
%%time
data_sorted = card_clean.copy()
#Cardnum, Merchnum
for groupbyvar in ['Cardnum', 'Merchnum']:
    data_sorted = data_sorted.sort_values(by = [groupbyvar, 'Date'])
    data_sorted_index = data_sorted.set_index('Date')
    for agg in ['mean', 'max', 'median', 'sum', 'count']: 
        for days in ['1d', '3d', '7d', '14d', '30d']:
            data_sorted[agg + '_' + groupbyvar + "_" + days]=getattr(data_sorted_index.groupby(groupbyvar)['Amount'].rolling(days),agg)().values
            data_sorted['Actual/' + agg + "_" + groupbyvar + "_" + days] = data_sorted['Amount']/data_sorted[agg + '_' + groupbyvar + "_" + days]
```

```python
data_sorted.sort_values(by = 'Recnum')
```

```python
data_sorted.info()
```

```python
for groupbyvar in ['Merchnum', 'Merch zip', 'Merch state']:
    data_sorted = data_sorted.sort_values(by = ['Cardnum',groupbyvar, 'Date'])
    data_sorted_index = data_sorted.set_index('Date')
    for agg in ['mean', 'max', 'median', 'sum', 'count']: 
        for days in ['1d', '3d', '7d', '14d', '30d']:
            data_sorted[agg + '_' + "Cardnum_" + groupbyvar + "_" + days] = getattr(data_sorted_index.groupby(['Cardnum',groupbyvar])['Amount'].rolling(days),agg)().values
            data_sorted['Actual/' + agg + "_" + "Cardnum_" + groupbyvar + "_" + days] = data_sorted['Amount']/data_sorted[agg + '_' + "Cardnum_" + groupbyvar + "_" + days]            
```

Days since Variables

```python
%%time
#Create days since variables
for groupbyvar in [['Cardnum'], ['Merchnum'], ['Cardnum', 'Merchnum'], ['Cardnum', 'Merch zip'], ['Cardnum', 'Merch state']]:
    sortCols = groupbyvar[:]
    sortCols.append('Date')
    data_sorted = data_sorted.sort_values(by = sortCols)
    if len(groupbyvar) == 1:
        data_sorted['Days_since_per_' + groupbyvar[0]] = data_sorted.groupby(groupbyvar)['Date'].apply(lambda x: (x - x.shift(1)).astype('timedelta64[D]')).fillna(365).values 
    else:
        data_sorted['Days_since_per_Cardnum_' + groupbyvar[1]] = data_sorted.groupby(groupbyvar)['Date'].apply(lambda x: (x -x.shift(1)).astype('timedelta64[D]')).fillna(365).values 
```

```python
data_sorted.info()
```

Velocity Expert Variables

```python
%%time
for groupbyvar1 in ['Cardnum', 'Merchnum']:
    data_sorted = data_sorted.sort_values(by = [groupbyvar1, 'Date'])
    data_sorted_index = data_sorted.set_index('Date')
    for agg1 in ['sum', 'count']: 
        for days1 in ['1d', '3d']:
            for groupbyvar2 in ['Cardnum', 'Merchnum']:
                for agg2 in ['sum', 'count']: 
                    for days2 in ['7d', '14d','30d']:
                        data_sorted[agg1 + '_' + groupbyvar1 + "_" + days1+"/"+agg2 + '_' + groupbyvar2 + "_" + days2]=getattr(data_sorted_index.groupby(groupbyvar1)['Amount'].rolling(days1),agg1)().values/getattr(data_sorted_index.groupby(groupbyvar2)['Amount'].rolling(days2),agg2)().values
                        
```
