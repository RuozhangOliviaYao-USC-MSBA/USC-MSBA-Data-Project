```python
card_origin = card_origin[["Recnum","Date"]]
```

```python
card_new=card.merge(card_origin,on='Recnum')
```

```python
card_new['Date'] = pd.to_datetime(card_new['Date']) # change date column to datetime data type

cols = card_new.columns.tolist()
cols = cols[-1:] + cols[:-1]
card_new = card_new[cols] 
```

```python
card_train = card_new.loc[(card_new['Date'] < '2010-11-01'),:]  # only use transactions before 1st Nov.
```

```python
card_train.dtypes
```

```python
card_train = card_train.reset_index(drop = True)
card_train['FraudLabel'] = card_train['Fraud']
card_train['randNum'] = np.random.randint(1, 1000000, card_train.shape[0])
# Create a fraud column and random number column for comparison
card_train.dtypes
```

## Calculate FDR Table for all the candidate variables

```python
# card_train_2.apply(lambda row: row.astype(str).str.contains('x').any(), axis=1)
```

```python
card_train.isnull().any()
card_train_2 = card_train.copy()

```

```python
Three_Percent = round(0.03 * len(card_train))
TotalFraud = card_train['Fraud'].sum()
FdrTable = pd.DataFrame(columns=['Variable', 'FDR'])
card_train_2['Fraud'] = card_train_2['Fraud'].astype('float64')
for column in card_train_2.iloc[:,5:]:
    FraudCaught = card_train_2[[column,'Fraud']].sort_values(by = [column], ascending = False)
    FDR = FraudCaught.iloc[0:Three_Percent,1:2].sum() / TotalFraud
    FdrTable = FdrTable.append({'Variable': column, 'FDR':float(FDR)}, ignore_index=True)


FdrTable = FdrTable.sort_values(by = 'FDR', ascending = False)
FdrTable
```

## Calculate KS for each candidate variable

```python
card_train_3 = card_train.copy()
KsTable = pd.DataFrame(columns=['Variable', 'KS'])
for column in card_train_3.iloc[:,5:]:
    KS = sps.ks_2samp(card_train_3.loc[card_train_3['Fraud'] == 0, column], card_train_3.loc[card_train_3['Fraud'] == 1, column])
    KsTable = KsTable.append({'Variable': column, 'KS':KS[0] }, ignore_index=True)
KsTable = KsTable.sort_values(by = 'KS', ascending = False)
KsTable
```

## Compile two tables and Save to an excel

```python
card_final=KsTable.merge(FdrTable,on='Variable')
```
